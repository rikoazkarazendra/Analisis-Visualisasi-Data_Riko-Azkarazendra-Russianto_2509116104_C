**Studi Kasus**

Menganalisis karakteristik parfum lokal Indonesia berdasarkan brand dan jenis aroma untuk mengetahui pola distribusi produk parfum di pasar. Melalui dataset ini, analisis dilakukan untuk memahami bagaimana persebaran jenis parfum berdasarkan merek dan aroma, serta melihat kecenderungan produk parfum lokal yang tersedia di Indonesia.

**Business Understanding**

  Business Understanding merupakan tahap awal dalam proses analisis data yang bertujuan untuk memahami permasalahan bisnis atau permasalahan data sebelum dilakukan pengolahan data lebih lanjut. Tahap ini terdiri dari empat bagian, yaitu Business Objective, Assess Situation, Analytic Goals, dan Project Plan.

**1. Business Objective**

  Dengan dari menganalisis dataset parfum Indonesia ini dapat mengetahui brand parfum yang paling banyak muncul dalam dataset, mengetahui jenis aroma parfum yang paling dominan (xdp,edp,edt), dan mengidentifikasi karakteristik umum parfum lokal Indonesia. Sehingga hasil analisis dapat memberikan gambaran mengenai tren parfum lokal di Indonesia.


**2. Asses Situation**

Pada dataset parfum Indonesia ini memiliki beberapa tantangan antara lain data berasal dari berbagai brand sehingga kemungkinan terdapat perbedaan format penulisan, terdapat kemungkinan data kosong pada beberapa kolom, sebagian besar kolom berupa data kategori (brand, aroma, tipe parfum) sehingga perlu dilakukan standarisasi sebelum dianalisis, dan dataset tidak memiliki target variabel tertentu, sehingga analisis yang dilakukan bersifat eksploratif.

**3. Analytic Goals**

Tujuan analisis data pada dataset ini adalah sebagai berikut:

1. Mengidentifikasi distribusi brand parfum lokal Indonesia.
2. Mengetahui jenis aroma (notes) yang paling sering muncul.
3. Menganalisis hubungan antara tipe parfum dengan jenis aroma.
4. Menyajikan insight berupa pola atau tren parfum lokal berdasarkan data.
5. Memberikan gambaran umum mengenai karakteristik parfum lokal Indonesia.

**4. Project Plan**

Rencana tahapan-tahapan pengerjaan proyek analisis dari dataset parfum Indonesia ini adalah sebagai berikut:

1. **Data Understanding**  
   - Memahami struktur dataset  
   - Mengidentifikasi jumlah baris dan kolom  
   - Mengetahui tipe data setiap kolom  

2. **Data Preparation**  
   - Membersihkan data (missing value dan duplikasi)  
   - Menyeragamkan penulisan kategori (brand, aroma, tipe parfum)  

3. **Visualization/Exploratory Data Analysis (EDA)**  
   - Menganalisis distribusi brand parfum  
   - Menganalisis jenis aroma parfum  
   - Visualisasi data menggunakan grafik

4. **Insight & Action**

    **Insight:**
    
    - Dataset didominasi oleh beberapa jenis aroma tertentu (misalnya fresh, sweet, woody, floral).
    - Setiap brand memiliki variasi produk parfum yang berbeda-beda.
    - Sebagian besar data berupa kategori (brand, aroma, tipe parfum).
    - Dataset dapat menggambarkan tren parfum lokal Indonesia.
    - Membandingkan jenis dan harga parfum.

    **Action:**

    - Melakukan data cleaning (missing value dan duplikasi).
    - Melakukan visualisasi distribusi brand dan jenis aroma.
    - Menganalisis hubungan antara tipe parfum, aroma, dan harga.
    - Mengambil insight dari hasil analisis.
