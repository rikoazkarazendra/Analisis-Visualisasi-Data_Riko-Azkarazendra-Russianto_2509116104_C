## **Cek Outliers**

Pada tahap ini, kita akan melihat nilai ekstrem (outliers) pada kolom yang bertipe data kan numerik. Hal ini dilakukan agar mendapatkan insight yang akurat sehingga dapat memperkuat analisis dan model yang akan kita bangun nantinya.

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

## **Data Preparation (Data Cleaning, Construct Data, Data Reduction)**

### **1. Penanganan Tipe Data**
Tipe data disesuaikan agar kolom numerik dapat dianalisis secara statistik dan kolom kategori dapat diproses sebagai teks.

```python
# Mengubah kolom numerik ke numerik
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df['size'] = pd.to_numeric(df['size'], errors='coerce')

# Mengubah kolom kategori menjadi object/string
cols_object = ['ID_Perfume','perfume','brand','concentrate','top notes','mid notes','base notes','situation','gender']
for col in cols_object:
    df[col] = df[col].astype(str)

df.info()
```

1. Kolom price berhasil diubah ke tipe numerik sehingga dapat digunakan dalam analisis statistik.

2. Kolom kategori diseragamkan menjadi tipe object agar mudah dikelompokkan.

3. Struktur dataset tetap berjumlah 1064 baris setelah proses konversi tipe data.

### **2. Penanganan Inconsistent Values**
Penyeragaman huruf dilakukan agar tidak terjadi perbedaan kategori akibat variasi penulisan seperti huruf besar dan kecil.

```python
df['brand'] = df['brand'].str.title()
df['concentrate'] = df['concentrate'].str.upper()
df['situation'] = df['situation'].str.title()
df['gender'] = df['gender'].str.title()

for col in ['brand','concentrate','situation','gender']:
    print(f"Unique values for {col}: {df[col].unique()}")
```

1. Nilai pada kolom brand diseragamkan agar tidak terjadi perbedaan kategori akibat variasi penulisan huruf.

2. Kolom concentrate dibersihkan dengan format huruf kapital untuk menjaga konsistensi data.

3. Kolom situation distandarisasi agar kategori dapat dianalisis dengan benar.

4. Kolom gender diseragamkan untuk menghindari duplikasi nilai yang sebenarnya sama.

### **3. Penanganan Missing Values**


```python
df.isnull().sum()
```

Karena tidak ada missing values pada dataset ini maka saya memilih untuk mengimputasi pada kolom numerik (price, size) menggunakan median agar distribusi data tidak terdistorsi oleh nilai ekstrem dan juga mengimputasi pada kolom kategorikal dengan nilai yang paling sering muncul (modus) agar tidak mengurangi jumlah data dan tetap konsisten dengan pola mayoritas data.

```python
# Imputasi kolom numerik dengan median
df['price'] = df['price'].fillna(df['price'].median())
df['size'] = df['size'].fillna(df['size'].median())

# Imputasi kolom kategorikal dengan modus
df['brand'] = df['brand'].fillna(df['brand'].mode()[0])
df['concentrate'] = df['concentrate'].fillna(df['concentrate'].mode()[0])
df['situation'] = df['situation'].fillna(df['situation'].mode()[0])
df['gender'] = df['gender'].fillna(df['gender'].mode()[0])
df['top notes'] = df['top notes'].fillna(df['top notes'].mode()[0])
df['mid notes'] = df['mid notes'].fillna(df['mid notes'].mode()[0])
df['base notes'] = df['base notes'].fillna(df['base notes'].mode()[0])

print("\nMissing values setelah imputasi:")
print(df.isna().sum())
```

1. Pengecekan missing values dilakukan untuk mengidentifikasi kolom yang memiliki data kosong.

2. Kolom numerik seperti price dan size diimputasi menggunakan median agar tidak terpengaruh oleh outlier.

3. Kolom kategorikal seperti brand, concentrate, situation, dan notes diimputasi menggunakan modus karena merupakan nilai yang paling sering muncul.

4. Proses imputasi dilakukan untuk menjaga jumlah data tetap lengkap dan menghindari penghapusan baris yang berpotensi menghilangkan informasi penting.

5. Setelah proses imputasi, seluruh kolom tidak ada memiliki missing values.

### **4. Penanganan Duplicate Values**
Data duplikat dihapus untuk mencegah perhitungan ganda dalam analisis.

```python
print("\nJumlah data duplikat:")
print(df.duplicated().sum())

df = df.drop_duplicates()

print("Jumlah data setelah hapus duplikat:")
print(df.shape)
```

Tidak ada data yang duplikat sehingga tidak ada data yang didrop/dihapus.

### **5. Penanganan Outliers (price & size)**
Outliers dibatasi menggunakan metode IQR agar distribusi data lebih stabil dan tidak ekstrem.

```python
columns_to_impute = ['price','size']

for col in columns_to_impute:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)

df[['price','size']].describe()
```

Berdasarkan output statistik deskriptif setelah penanganan outliers menggunakan metode IQR, nilai minimum dan maksimum pada kolom price dan size berada dalam rentang yang lebih wajar. Hal ini menunjukkan bahwa nilai ekstrem yang sebelumnya berpotensi memengaruhi analisis telah dibatasi sehingga distribusi data menjadi lebih stabil.

### **6. Construct Data (Membuat Kolom Baru)**
Kolom baru dibuat untuk mengelompokkan parfum berdasarkan harga dan ukuran agar analisis lebih mudah.

```python
# Kolom kategori harga
def kategori_harga(harga):
    if harga < 200000:
        return 'Murah'
    elif harga <= 500000:
        return 'Sedang'
    else:
        return 'Mahal'
df['kategori_harga'] = df['price'].apply(kategori_harga)

# Kolom kategori ukuran
def kategori_ukuran(size):
    if size < 30:
        return 'Kecil'
    elif size <= 60:
        return 'Sedang'
    else:
        return 'Besar'
df['kategori_ukuran'] = df['size'].apply(kategori_ukuran)

df[['price','kategori_harga','size','kategori_ukuran']].head(10)
```

1. Kategori harga dibagi menjadi Murah, Sedang, dan Mahal berdasarkan rentang nilai harga parfum.

2. Kategori ukuran dibagi menjadi Kecil, Sedang, dan Besar berdasarkan volume botol parfum.

3. Hasil konstruksi data menghasilkan variasi kategori yang lebih representatif dibandingkan sebelumnya yang hanya menghasilkan satu kategori.

### **7. Data Reduction (Menghapus Kolom Tidak Relevan)**
Kolom yang tidak berpengaruh pada analisis dihapus untuk menyederhanakan dataset.
